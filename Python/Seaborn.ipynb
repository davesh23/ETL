# BoxPlot

```python
#a box plot are used to show the distribution of catogerial data.
#the show he distribution of quntitative data in away that facilitates between variable or across levels of a categorial variable
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```

```python
#Do smokker eat more
sns.boxplot(x = 'day', y='total_bill', hue="smoker", data= tips, palette=  "coolwarm")
```

```python
#Female give less tips
sns.boxplot(x = 'day', y = 'size', hue = "sex", data = tips, palette = "coolwarm")
```
